## III 

```python
from geopy.distance import geodesic
from geopy.geocoders import ArcGIS

    
# Function to calculate accuracy
def calculate_accuracy(row):
    temp_str = row['Coordinates']
    given_coords = tuple(map(float, temp_str.split(',')))
    address_coords = geocode_address(row['Address'])

    # Calculate the distance in meters
    distance = geodesic(given_coords, address_coords).meters

    # Calculate accuracy
    if distance == 0:
        return 100.0  # Exact match
    else:
        accuracy = max(0, 100 - (distance / 10))
        return round(accuracy, 2)


# Function to simulate geocoding
def geocode_address(address):
    nom = ArcGIS()
    coordinate = nom.geocode(address)
    rounded_tuple = tuple(round(value, 5) for value in coordinate[1])
    return rounded_tuple 


# Apply the calculate_accuracy function to each row
df2['GeoAccuracy'] = df2.apply(calculate_accuracy, axis=1)
```

```python
import pandas as pd
import plotly.express as px
import re

def classify_and_plot(year):
    # Load the CSV data
    data = pd.read_csv('MH-Vehicle-Reg.csv', delimiter=';')

    # Filter data for the specified year
    year_data = data[data['Year'] == year]

    # Define vehicle categories
    private_vehicles = ['Motor Cycles', 'Scooters', 'Moped', 'Cars', 'Jeeps', 'Stn. Wagons', 'Private Service Vehicles']
    commercial_vehicles = ['Taxis meter fited', 'Luxury /Turist Cabs/', 'Auto-rikshaws', 
                            'Stage carriages', 'Contract carriages /Mini Bus', 'School Buses', 
                            'Ambulances', 'Articulated/Multi.', 'Trucks & Lorries', 'Tanker', 
                            'Delivery Van (4 wheelers)', 'Delivery Van (3 wheelers)', 'Tractors', 'Trailors']
    others = ['Others'] 
    hmvs = ['Stage carriages', 'Contract carriages /Mini Bus', 'School Buses', 'Trucks & Lorries', 'Tanker', 'Tractors', 'Trailors']
    lmvs = ['Taxis meter fited', 'Luxury /Turist Cabs/', 'Auto-rikshaws','Ambulances', 'Articulated/Multi.','Delivery Van (4 wheelers)', 'Delivery Van (3 wheelers)']
    # Calculate the total number of vehicles in each category
    private_total = year_data[private_vehicles].sum().sum()
    commercial_total = year_data[commercial_vehicles].sum().sum()
    other_total = year_data[others].sum().sum() 
    lmvs_total = year_data[lmvs].sum().sum() 
    hmvs_total = year_data[hmvs].sum().sum() 

    private_all_category_total = year_data[private_vehicles].sum()
    private_all_category_names = list(private_all_category_total.keys())
    p_len = len(private_all_category_names)
    
    others_all_category_total = year_data[others].sum()
    others_all_category_names = list(others_all_category_total.keys())
    o_len = len(others_all_category_names)
    
    lmvs_all_category_total = year_data[lmvs].sum()
    lmvs_all_category_names = list(lmvs_all_category_total.keys())
    l_len = len(lmvs_all_category_names)
    
    hmvs_all_category_total = year_data[hmvs].sum()
    hmvs_all_category_names = list(hmvs_all_category_total.keys())
    h_len = len(hmvs_all_category_names)
    
    total = private_total + commercial_total + other_total

    #define dictionary for sunburst
    data = dict(
    categories = ['Vehicles', 'Commercial', 'Private', 'OthersVehicles', 'LowMotor', 'HeavyMotor'],
    parent = ['', 'Vehicles', 'Vehicles', 'Vehicles', 'Commercial', 'Commercial'],
    value = [total, commercial_total, private_total, other_total, lmvs_total, hmvs_total])

    #define loops to append sub-category data to dictionary
    for x in range(p_len):
        data['categories'].append(private_all_category_names[x])
        data['parent'].append('Private')
        data['value'].append(private_all_category_total[x])
        
    for x in range(o_len):
        data['categories'].append(others_all_category_names[x])
        data['parent'].append('OthersVehicles')
        data['value'].append(others_all_category_total[x])

    for x in range(l_len):
        data['categories'].append(lmvs_all_category_names[x])
        data['parent'].append('LowMotor')
        data['value'].append(lmvs_all_category_total[x])
        
        
    for x in range(h_len):
        data['categories'].append(hmvs_all_category_names[x])
        data['parent'].append('HeavyMotor')
        data['value'].append(hmvs_all_category_total[x])

    
    fig = px.sunburst(data,
                  names = 'categories',
                  parents = 'parent',
                  values = 'value',
                  color_discrete_sequence = ['teal'], #olive,orchid,tan
                  branchvalues = 'total')
    
    fig.update_traces(textinfo="label+value+percent parent")
    fig.update_traces(insidetextorientation='auto')
    fig.update_layout(height=900, width=900,showlegend=True)

    
    fig.show() 
    


# Regular expression pattern for the expected format YYYY-YYYY
year_range_pattern = re.compile(r'^20(01|02|03|04|05|06|07|08|09|10|11|12|13|14|15|16|17|18)-20(02|03|04|05|06|07|08|09|10|11|12|13|14|15|16|17|18)$')

while True:
    user_input = input("Enter a year range in the format YYYY-YYYY (2001-2018): ")

    # Check if the input matches the expected pattern
    if year_range_pattern.match(user_input):
        start_year, end_year = map(int, user_input.split('-'))
        
        # Check if the years are consecutive and within the range 2001-2018
        if start_year + 1 == end_year and 2001 <= start_year <= 2018:
            break
        else:
            print("Years must be consecutive and in the range 2001-2018. Please try again.")
    else:
        print("Invalid format. Please enter a year range in the format YYYY-YYYY (2001-2018).")


print(f"Plot for the year range: {start_year}-{end_year}")

classify_and_plot(user_input)
```
